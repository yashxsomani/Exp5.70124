```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generating random data
data = {'Day': np.arange(1, 11), 'Value': np.random.randint(10, 50, 10)}

# Creating a DataFrame
df = pd.DataFrame(data)

# Plotting the data
df.plot(x='Day', y='Value', kind='line', marker='o', title="Random Values Over 10 Days")

# Show the plot
plt.show()
```
